# IF Elif Else #

# Int #

```python
score = int (input('provide user score = '))
print(f'score = {score} , type of score = {type(score)}')

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")    
else:
    print("Grade: D or below")
```

# String #

```python
password = input('provide user score ="" ')
print(f'password = {password} , type of password = {type(password)}')
if len(password) < 5:
    print("Status: Weak")
elif "123" in password:
    print("Status: Common/Predictable")
else:
    print("Status: Strong")
```

# List #

```python
cart = ["apple", "banana", "orange", "grape", "mango", "peach"]
print(f'cart = {cart} , type of cart = {type(cart)}')

if len(cart) == 0:
    print("Your cart is empty.")
elif len(cart) > 5:
    print("Bulk discount applied!")
elif len(cart) > 2:    
    print("You have a medium cart.")
else:
    print(f"You have {len(cart)} items in your cart.")
```

# Dictionary #

```python
user = {"username": "dev_user", "role": "editor", "experience": 3}
print(f'user = {user} , type of user = {type(user)}')

if user["role"] == "admin":
    print("Full access granted.")
elif user["role"] == "editor":
    print("Partial access: Can edit but not delete.")
elif user["experience"] < 1:
    print("Limited access: New user.")  
else:
    print("Read-only access.")
```

# Set #

```python
shared_hobbies = {"cycling", "reading"}
print(f'shared_hobbies = {shared_hobbies} , type of shared_hobbies = {type(shared_hobbies)}')

if not shared_hobbies:
    print("No common interests found.")
elif "coding" in shared_hobbies:
    print("You both love tech!")

else:
    print(f"You share {len(shared_hobbies)} hobbies.")
```

# Tuple #

```python
location = (0, 0)
print(f'location = {location} , type of location = {type(location)}')

if location == (0, 0):
    print("You are at the Null Island.")
elif location[0] > 0:
    print("You are in the Northern .")
else:
    print("You are in the Southern .")
```
